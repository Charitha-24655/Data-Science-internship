```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/cheli/OneDrive/Desktop/sampledata.csv")
```

```python
df
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Name'], df['Age'], marker='o', linestyle='-')
plt.title('Name vs. Age')
plt.xlabel('Name')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
grouped_salary = df.groupby('Department')['Salary'].mean()
plt.figure(figsize=(6, 4))
plt.bar(grouped_salary.index, grouped_salary.values)
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Salary'])
plt.title('Scatter Plot of Salary vs. Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
