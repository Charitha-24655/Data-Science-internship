```python
df = df.replace(" ?", np.nan)
df.isnull().sum()
```

```python
#Handle Missing Values
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
#2. Scaling Techniques
#Numerical Columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

#a) Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standard_scaled = df.copy()
df_standard_scaled[num_cols] = scaler.fit_transform(df[num_cols])
print(df_standard_scaled.head())
```

```python
#b) Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_minmax_scaled = df.copy()
df_minmax_scaled[num_cols] = scaler.fit_transform(df[num_cols])
print(df_minmax_scaled.head())
```

```python
# 3. Encoding Techniques
#One-Hot Encoding (for <5 categories)
df = pd.get_dummies(df, columns=['sex','race','relationship'])
print(df.columns)
```

```python
print(df.columns)
```

```python
#Label Encoding (for >5 categories)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = ['workclass','occupation','native_country','education','marital_status']

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
print(df.head())
```

```python
Pros and Cons of Encoding Methods
One-Hot Encoding

Pros
  No ordinal relationship created
  Works well for nominal data
Cons
  Increases number of columns
  cause high dimensionality

Label Encoding
Pros
  Simple and fast
  Does not increase number of columns
Cons
  May introduce false order between categories
```

```python
#4. Feature Engineering
#Feature 1: Capital Difference
df['capital_diff'] = df['capital_gain'] - df['capital_loss']
print(df[['capital_gain','capital_loss','capital_diff']].head())
```

```python
#Feature 2: Work Intensity
df['work_intensity'] = df['hours_per_week'] * df['education_num']
print(df[['hours_per_week','education_num','work_intensity']].head())
```

```python
#5. Feature Transformation
#Log Transformation for Skewed Feature
df['log_capital_gain'] = np.log1p(df['capital_gain'])
print(df[['capital_gain','log_capital_gain']].head())
```
