```python
import pandas as pd
# load the data
data = pd.read_csv("Cardiotocographic.csv")
# Display the first few rows of the dataset
data.head()
```

```python
# Display the information about the dataset
print(f"Data Info: {data.info()}")
# count the number of missing values in each column
# This will help identify if there are any columns with missing data
print(f"Count of Null values:\n {data.isnull().sum()}")
```

```python
# Calculate the mean and standard deviation of the dataset
# This will give a quick overview of the central tendency and spread of the data
data.describe()
```

```python
# Fill missing values with the mean of each column
# This is a common method to handle missing data, especially when the amount of missing data is small
# It helps to maintain the dataset's integrity without losing too much information
data_cleaned = data.fillna(data.mean())
display(data_cleaned.isnull().sum())
```

```python
# Data Visualization
# Importing necessary libraries for visualization
import matplotlib.pyplot as plt

# Plotting histograms for each column in the dataset
# This will help visualize the distribution of each feature
data_cleaned.hist(figsize=(15, 10),color = "skyblue")
plt.tight_layout()
plt.show()
```

```python
# Scatter plots to visualize relationships between pairs of features
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
# scatter plots for some correlated columns
sns.scatterplot(data=data_cleaned, x='ASTV', y='MSTV',color = 'red')
plt.title('Scatter plot of ASTV vs MSTV')
plt.show()

# scatter plot of ALTV vs MLTV
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data_cleaned, x='ALTV', y='MLTV',color = 'skyblue')
plt.title('Scatter plot of ALTV vs MLTV')
plt.show()

# scatter plot of Width vs MSTV
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data_cleaned, x='Width', y='MSTV',color = 'green')
plt.title('Scatter plot of Width vs MSTV')
plt.show()
```

```python
# Detect and treat outliers using IQR
# Calculate IQR for each column and identify outliers
# This will help in understanding the spread of the data and identifying potential outliers

import numpy as np
import pandas as pd

outlier_summary = []

for col in data_cleaned.columns:
    q1 = np.percentile(data_cleaned[col], 25) # 25th percentile
    q3 = np.percentile(data_cleaned[col], 75) # 75th percentile
    IQR = q3 - q1  # Interquartile range
    UW = q3 + (1.5 * IQR) # Upper Whisker
    LW = q1 - (1.5 * IQR) # Lower Whisker

    # Count outliers
    num_upper = (data_cleaned[col] > UW).sum() # Count of upper outliers
    num_lower = (data_cleaned[col] < LW).sum() # Count of lower outliers

    # Replace outliers with the whisker values
    # This will help in treating the outliers without removing them from the dataset
    # It ensures that the data remains within a reasonable range
    data_cleaned[col] = np.where(data_cleaned[col] > UW, UW,
                        np.where(data_cleaned[col] < LW, LW, data_cleaned[col]))

    # Collect summary
    # This will help in understanding the spread of the data and identifying outliers
    outlier_summary.append({
        "Column": col,
        "Quartile 1": q1,
        "Quartile 3": q3,
        "IQR": IQR,
        "Lower Whisker": LW,
        "Upper Whisker": UW,
        "Upper Outliers": num_upper,
        "Lower Outliers": num_lower
    })

# displaying the summary table in df
outlier_df = pd.DataFrame(outlier_summary)
display(outlier_df)
```

```python
# Before vs After Outlier Treatment
import matplotlib.pyplot as plt
import seaborn as sns

# ---- BEFORE Treatment ----
fig, axes = plt.subplots(2, 5, figsize=(18,8))
axes = axes.flatten()

# Plotting boxplots for each column to visualize outliers
# This will help identify any outliers in the dataset
for i, col in enumerate(data.columns[:10]):  # first 10 columns
    sns.boxplot(x=data[col], ax=axes[i], color="red") # red for before treatment
    axes[i].set_title(f"{col} (Before Treatment)") # title for each subplot

# Set the title for the entire figure -- y offset to avoid overlap
plt.suptitle("Boxplots (Before Outlier Treatment)", fontsize=16, y=1.02)
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()

# ---- AFTER Treatment ----
fig, axes = plt.subplots(2, 5, figsize=(18,8))
axes = axes.flatten()

for i, col in enumerate(data_cleaned.columns[:10]):  # first 10 columns
    sns.boxplot(x=data_cleaned[col], ax=axes[i], color="green") # green for after treatment
    axes[i].set_title(f"{col} (After Treatment)")

# suptitle for each subplot
plt.suptitle("Boxplots (After Outlier Treatment)", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
