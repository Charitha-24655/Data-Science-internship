```python
numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# append
numbers.append(60)
print("After append:", numbers)

# extend
numbers.extend([70, 80])
print("After extend:", numbers)

# insert
numbers.insert(2, 25)
print("After insert:", numbers)

# remove
numbers.remove(40)
print("After remove:", numbers)

# pop
numbers.pop()
print("After pop:", numbers)

# index
print("Index of 30:", numbers.index(30))

# count
print("Count of 20:", numbers.count(20))

# sort
numbers.sort()
print("After sort:", numbers)

# reverse
numbers.reverse()
print("After reverse:", numbers)

# clear
numbers.clear()
print("After clear:", numbers)
```

```python
import pandas as pd
import numpy as np

# Create DataFrame with 10 rows
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva",
             "Frank", "Grace", "Helen", "Ivy", "Alice"],
    "Age": [25, 30, 35, 28, 22,
            40, 29, 31, 27, 25],
    "City": ["Mumbai", "Delhi", "Chennai", "Kolkata", "Pune",
             "Delhi", "Mumbai", "Chennai", "Pune", "Mumbai"],
    "Salary": [50000, 60000, np.nan, 55000, 48000,
               70000, 52000, np.nan, 51000, 50000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

# -----------------------------
# 1️⃣ Info
print("\nDataFrame Info:\n")
print(df.info())

# -----------------------------
# 2️⃣ Describe
print("\nDataFrame Description:\n")
print(df.describe())

# -----------------------------
# 3️⃣ Select 'Name' and 'City'
print("\nSelected Columns (Name & City):\n")
print(df[["Name", "City"]])

# -----------------------------
# 4️⃣ Drop 'City' column
df = df.drop("City", axis=1)
print("\nAfter Dropping 'City' Column:\n")
print(df)

# -----------------------------
# 5️⃣ Fill missing Salary with mean
mean_salary = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(mean_salary)

print("\nAfter Filling Missing Salary with Mean:\n")
print(df)

# -----------------------------
# 6️⃣ Remove duplicate rows
df = df.drop_duplicates()

print("\nAfter Removing Duplicates:\n")
print(df)
```
