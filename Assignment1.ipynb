```python
n = int(input())
is_prime = True   # assume prime initially

if n <= 1:
    is_prime = False
else:
    for i in range(2, n):
        if n % i == 0:   # divisible ayithe
            is_prime = False
            break

if is_prime:
    print(n, "is a Prime number")
else:
    print(n, "is NOT a Prime number")
```

```python
import random
a = random.randint(1, 10)
b = random.randint(1, 10)

print("Multiply:", a, "x", b)
c = int(input("Enter your answer: "))

if c == a * b:
    print("✅ Correct!")
else:
    print("❌ Wrong! Correct answer is:", a * b)
```

```python
choice = input("Do you want EVEN or ODD squares?").lower()

for num in range(100, 201):
    if choice == "even" and num % 2 == 0:   # even numbers
        print(num, num * num)
    elif choice == "odd" and num % 2 != 0:  # odd numbers
        print(num, num * num)
```

```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300
discount_rate = 0.10

# Patient details
days_stayed = 4
number_of_tests = 3
medicine_charges = 2400

# Calculations
room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_cost * number_of_tests

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if applicable
if total_bill > 10000:
    discount = total_bill * discount_rate
else:
    discount = 0

final_bill = total_bill - discount

# Output
print("Room Charges:", room_charges)
print("Doctor Fee:", doctor_fee)
print("Lab Charges:", lab_charges)
print("Medicine Charges:", medicine_charges)
print("Total Bill before Discount:", total_bill)
print("Discount:", discount)
print("Final Bill to be Paid:", final_bill)
```
