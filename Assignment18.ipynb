```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("sonardataset.csv")
print(df.head())
```

```python
#Dataset Overview
print("Shape of dataset:", df.shape)
print("\nData Info")
print(df.info())

print("\nClass Distribution")
print(df['Y'].value_counts())
```

```python
#Check Missing Values
print(df.isnull().sum())
```

```python
#Separate Features and Target
X = df.drop('Y', axis=1)
y = df['Y']
```

```python
#Encode Target Labels
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
#Normalize the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Training samples:", X_train.shape)
print("Test samples:", X_test.shape)
```

```python
# 2.Model Implementation (Basic ANN)
#Input Layer (60 neurons)
#Hidden Layer (32 neurons)
#Hidden Layer (16 neurons)
#Output Layer (1 neuron)
model = Sequential()

model.add(Dense(32, input_dim=60, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2
)

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)
```

```python
# 3.Hyperparameter Tuning
model_tuned = Sequential()

model_tuned.add(Dense(64, input_dim=60, activation='relu'))
model_tuned.add(Dense(32, activation='relu'))
model_tuned.add(Dense(16, activation='relu'))
model_tuned.add(Dense(1, activation='sigmoid'))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model_tuned.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_tuned.fit(
    X_train,
    y_train,
    epochs=150,
    batch_size=8,
    validation_split=0.2
)
```

```python
# 4.Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Classification Report
print(classification_report(y_test, y_pred))
```

```python
#Confusion Matrix
print(confusion_matrix(y_test, y_pred))
```

```python
Accuracy

Accuracy measures the proportion of correctly classified samples.
Accuracy= (TP+TN)/(TP+TN+FP+FN)
It shows the overall performance of the model.

Precision

Precision measures how many predicted mines are actually mines.
Precision= TP/(TP+FP)
High precision means few false alarms.

Recall

Recall measures how many actual mines were correctly detected.
Recall= TP/(TP+FN)
High recall is important because missing a mine can be dangerous.

F1 Score

F1-score balances precision and recall.
F1=2Г—[(Precision+Recall)/(PrecisionГ—Recall)
It is useful when both false positives and false negatives matter.
```
