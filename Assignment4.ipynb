```python

import numpy as np
import pandas as pd
import scipy.stats as st

# Load datasets
patients = pd.read_csv("Patient_Data.csv")
billing = pd.read_csv("Billing_Data.csv")

# Check dataset summary
print(patients.info())

# Select columns relevant for billing
billing_cols = patients[['PatientID','Department','Doctor','BillAmount']]

# Drop administrative columns
patients_clean = patients.drop(columns=['ReceptionistID','CheckInTime'], errors='ignore')

# Department-wise total bill
dept_total = patients_clean.groupby('Department')['BillAmount'].sum()
print(dept_total)

# Remove duplicate patient records
patients_unique = patients_clean.drop_duplicates(subset='PatientID')

# Fill missing bill amounts with mean
mean_bill = patients_unique['BillAmount'].mean()
patients_unique['BillAmount'] = patients_unique['BillAmount'].fillna(mean_bill)

# Merge patient and billing datasets
merged_data = pd.merge(patients_unique, billing, on='PatientID', how='left')

```

```python
# New patients for current week
new_patients = pd.DataFrame({
    'PatientID':[301,302],
    'Department':['Cardiology','Neurology'],
    'Doctor':['Dr.Kumar','Dr.Singh'],
    'BillAmount':[4500,5200]
})

# Row-wise concatenation
updated_data = pd.concat([merged_data, new_patients], axis=0, ignore_index=True)

print(updated_data.head())
```

```python
extra_columns = pd.DataFrame({
    'InsuranceCovered':['Yes','No','Yes','Yes'],
    'FinalAmount':[4300,5200,6000,4800]
})

# Column-wise concatenation
final_dataset = pd.concat([updated_data, extra_columns], axis=1)

print(final_dataset.head())
```
